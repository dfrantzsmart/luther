## Encode isNotMissing
Our data has missing values.  
We will create a new DF of the same shape as our original DF.  
The new DF will contain a 1 if the corresponding cell in the original DF has a value, and 0 if it is missing a value.  
We will then replace all missing values in the original DF with 0s.  
Finally, we will concatenate the two DFs, by columns, into one unified DF.  

```python
# define function to check if DF cell has a value
def isNotMissing(x):
    out = np.float64(not pd.isnull(x))
    return out
```

```python
# calculate how many values in the dataframe are missing
list(cleanData.applymap(isNotMissing).values.flatten()).count(0)
```

```python
# calculate how many values in the dataframe are NOT missing
list(cleanData.applymap(isNotMissing).values.flatten()).count(1)
```

```python
# create a DF that tells us if the cell has a value
cleanDataNotMissing = cleanData.applymap(isNotMissing).copy()
#cleanDataNotMissing
```

```python
# define new names for the columns
notMissingColName = cleanDataNotMissing.columns.values
notMissingColName = [ 'notMissing' + x for x in notMissingColName.tolist() ]
#notMissingColName
```

```python
# set the column names to the new ones
cleanDataNotMissing.columns = ['Date'] + (notMissingColName[1::])
#cleanDataNotMissing
```

```python
# replace the Date column with actual Dates, because we will use 
# as the index
cleanDataNotMissing['Date'] = cleanData['Date']
#cleanDataNotMissing
```

```python
# define the columns that will be used in features
featureCols = ['Volume', 'DateDelta', 'Return', 'HighLow', 'Market Cap (intraday) 5', \
              'Enterprise Value 3', 'Trailing P/E ', 'Forward P/E 1', 'PEG Ratio (5 yr expected) 1', \
              'Price/Sales (ttm)', 'Price/Book (mrq)', 'Enterprise Value/Revenue 3', 'Enterprise Value/EBITDA 6' ]

# # This is for using Price levels, rather than returns
# featureCols = ['Volume', 'DateDelta', 'Adj Close', 'HighLow', 'Market Cap (intraday) 5', \
#               'Enterprise Value 3', 'Trailing P/E ', 'Forward P/E 1', 'PEG Ratio (5 yr expected) 1', \
#               'Price/Sales (ttm)', 'Price/Book (mrq)', 'Enterprise Value/Revenue 3', 'Enterprise Value/EBITDA 6' ]
```

```python
# define the columns for the "not missing" DF that will be added
# to the features
notMissingFeatureCols = [ 'notMissing' + x for x in featureCols ]
#notMissingFeatureCols
```

```python
# define the target DF

allTargets = cleanData.set_index('Date')['FutureReturn'].copy()

# # This is for using Price levels, rather than returns
# allTargets = cleanData.set_index('Date')['FuturePrice'].copy()

#allTargets
```

```python
# define the input DF
allFeatures = cleanData.set_index('Date')[featureCols].copy()
allFeatures
```

```python
# define the missing features DF
notMissingAllFeatures = cleanDataNotMissing.set_index('Date')[notMissingFeatureCols].copy()
notMissingAllFeatures
```

```python
# make sure that the features DF, and not-missing DF have 
# the same shape
notMissingAllFeatures.shape == allFeatures.shape
```

```python
# replace missing values with zeros in the features DF
allFeatures.fillna(0.0, inplace=True)
allFeatures
```

```python
# combine the features DF, and not-missing DF, into one DF
finalAllFeatures = pd.concat([allFeatures, notMissingAllFeatures], axis=1)
finalAllFeatures
```

## Split Data
We split our data into Training and Dev sets. 

```python
# split data into training set 
trainTgt = allTargets[ allTargets.index < sections[-2] ].copy()
trainInp = finalAllFeatures[ finalAllFeatures.index < sections[-2] ].copy()
trainInpControl = trainInp['Return'].copy()
# trainInpControl = trainInp['Adj Close'].copy()
```

```python
# split data into dev set
devTgt = allTargets[ (allTargets.index > sections[-2]) & (allTargets.index < sections[-1]) ].copy()
devInp = finalAllFeatures[ (finalAllFeatures.index > sections[-2]) & (finalAllFeatures.index < sections[-1]) ].copy()
devInpControl = devInp['Return'].copy()
# devInpControl = devInp['Adj Close'].copy()
```

## Plot AFC
Lets see if there is any significant serial correlation for returns

```python
x_train = trainInpControl.values
y_train = trainTgt.values
```

```python
x_train_acfY = acf(x_train,fft=True)
acfX = np.arange(1,len(x_train_acfY)+1,1)
twoStDs = np.std(x_train_acfY)*2
plt.stem(x_train_acfY)
plt.axhline(0, color='black', linewidth=2)
plt.axhline(twoStDs, color='red', linewidth=1, linestyle='--')
plt.axhline(-twoStDs, color='red', linewidth=1, linestyle='--')
```

There is not. I bet ploting current returns against future returns will give us a big circular blob.

```python
sns.mpl.rc('figure',figsize=(10, 10))
sns.scatterplot(x=x_train, y=y_train, alpha=0.1)
```

Yes it does.  
An AR model is NOT appropriate for modeling this data.

## Create Simple Model
Even though an AR model is not appropriate, we will still construct one.

```python
x_train = trainInpControl.values
y_train = trainTgt.values

y_model = y_train
x_model = np.concatenate((np.ones(len(x_train)).reshape(-1,1),x_train.reshape(-1,1)), axis=1)
regAR1ControlModel = sm.OLS(y_model, x_model).fit()
regAR1ControlModel.summary()
```

As expected, this is a terrible model, and does not explain much.

## Create Complex Model
We should expect multi-colinearity because the Key Statistics share a lot of variables

```python
x_train = trainInp.values
y_train = trainTgt.values

y_model = y_train
x_model = np.concatenate((np.ones(len(x_train)).reshape(-1,1),x_train), axis=1)
regAR1ExperimentModel = sm.OLS(y_model, x_model).fit()
regAR1ExperimentModel.summary()
```

This is also a terrible model, but it appears to explain more than the simple model.

Lets get the Mean Squared Error of each model.

```python
x_trainCtrl = trainInpControl.values
x_trainCtrl= np.concatenate((np.ones(len(x_trainCtrl)).reshape(-1,1),x_trainCtrl.reshape(-1,1)), axis=1)
x_trainExpmt = trainInp.values
x_trainExpmt = np.concatenate((np.ones(len(x_trainExpmt)).reshape(-1,1),x_trainExpmt), axis=1)
y_train = trainTgt.values
predtrainCtrl = regAR1ControlModel.predict(x_trainCtrl)
predtrainExpmt = regAR1ExperimentModel.predict(x_trainExpmt)
trainCtrl_MSE = ((predtrainCtrl - y_train)**2).mean()
trainExpmt_MSE = ((predtrainExpmt - y_train)**2).mean()
```

```python
trainCtrl_MSE
```

```python
trainExpmt_MSE
```

# Make Predictions with Dev Data
Let's use some out-of-sample data to make predictions with the Simple and Complex models.

```python
x_devCtrl = devInpControl.values
x_devCtrl= np.concatenate((np.ones(len(x_devCtrl)).reshape(-1,1),x_devCtrl.reshape(-1,1)), axis=1)
x_devExpmt = devInp.values
x_devExpmt = np.concatenate((np.ones(len(x_devExpmt)).reshape(-1,1),x_devExpmt), axis=1)
y_dev = devTgt.values
```

```python
predDevCtrl = regAR1ControlModel.predict(x_devCtrl)
predDevExpmt = regAR1ExperimentModel.predict(x_devExpmt)
```

```python
devCtrl_MSE = ((predDevCtrl - y_dev)**2).mean()
devExpmt_MSE = ((predDevExpmt - y_dev)**2).mean()
```

```python
devCtrl_MSE
```

```python
devExpmt_MSE
```

Looks like the Complex model performed worse than the Simple model with new data!

Lets plot a Density plot of the squared residuals for each model:

```python
devCtrl_SE = ((predDevCtrl - y_dev)**2)
devExpmt_SE = ((predDevExpmt - y_dev)**2)
SqRtErrDf = pd.DataFrame()
SqRtErrDf['Control'] = devCtrl_SE
SqRtErrDf['Experiment'] = devExpmt_SE
```

```python
sns.mpl.rc('figure',figsize=(10, 5))
#sns.scatterplot(x=devInpControl.values, y=y_dev, alpha=0.05)
#sns.lineplot(x=devInpControl.values, y=devCtrl_SE, color='black')
plt.xlim(-0.002, 0.004)
#plt.xlim(0, 5000)
sns.kdeplot(devExpmt_SE)
sns.kdeplot(devCtrl_SE, color='red')
```

The distributions look pretty much identical... idk if I could say they are different with confidence. 

I'll run a one-way ANOVA to see what it says: 

```python
scipy.stats.f_oneway(devCtrl_SE.flatten(), devExpmt_SE.flatten(),)
```

The ANOVA says the same thing: we can't say with high confidence that there is a difference. 

If choosing between the Simple and Complex model, it's better to use the Simple. 

## Create Neural Network Model
2 layer:
L1: 26 nodes; Leaky-ReLu activation
L2: 1 nodes; ReLu activation

```python
x_train = trainInp.values
#x_train
```

```python
y_train = trainTgt.values.reshape(-1,1)
#y_train
```

```python
model = kr.Sequential([
    kr.layers.Dense(26, activation=tf.nn.leaky_relu),
    kr.layers.Dense(1, activation=tf.nn.relu),
])
```

```python
model.compile(optimizer=tf.optimizers.Adam(),
              loss='mse',)

model.fit( x_train, y_train, epochs=1)
```

## Make Predictions with Dev Data
We'll use out-of-sample observations to make predictions with the NN

```python
x_dev = devInp.values
```

```python
y_dev = devTgt.values.reshape(-1,1)
```

```python
predDev = model.predict(x_dev)
```

```python
mse = ((predDev-y_dev)**2).mean()
mse
```

Interesting... The NN has a lower Mean Square Error than the other two models...

Let's plot the density of the residuals for all 3 models:

```python
nnDev_SE = (predDev-y_dev)**2
```

```python
sns.mpl.rc('figure',figsize=(10, 5))
plt.xlim(-0.002, 0.004)

sns.kdeplot(devExpmt_SE)
sns.kdeplot(devCtrl_SE, color='red')
sns.kdeplot(nnDev_SE.flatten(), color='green')
```

This actually looks like it has a relatively significant amount more of residuals centered around 0. 
It implies that the NN model is better than the other two....

Lets run a one-way ANOVA on all 3 model residuals to see what it says:

```python
scipy.stats.f_oneway(devCtrl_SE.flatten(), devExpmt_SE.flatten(), nnDev_SE.flatten())
```

That's a pretty high F-Score! 

I think the evidence suggest that the NN is indeed more accurate than the other two models.  
Of the 3, it is the best to use.
